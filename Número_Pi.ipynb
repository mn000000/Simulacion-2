```python
import random
import matplotlib.pyplot as plt

def approximate_pi(num_samples):
    inside_circle = 0
    total_points = num_samples
    y_values = []

    for i in range(num_samples):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)

        distance = x**2 + y**2

        if distance <= 1:
            inside_circle += 1
        pi_estimate = (inside_circle / (i + 1)) * 4
        y_values.append(pi_estimate)

    return y_values

# Definimos el número de muestras (mientras más muestras, mayor precisión)
num_samples = 1000
pi_estimations = approximate_pi(num_samples)
print(pi_estimations)

# Gráfico de la aproximación de Pi a través de las muestras
plt.figure(figsize=(12, 6))

# Gráfico de línea
plt.subplot(1, 2, 1)
plt.plot(range(1, num_samples + 1), pi_estimations, color='blue')
plt.title('Aproximación de Pi a través de la simulación de Monte Carlo')
plt.xlabel('Número de muestras (N)')
plt.ylabel('Aproximación de Pi')
plt.axhline(y=3.14159, color='red', linestyle='--', label='Valor real de Pi')
plt.legend()
plt.grid(True)
```
